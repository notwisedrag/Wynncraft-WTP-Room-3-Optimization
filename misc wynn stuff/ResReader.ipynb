```python
import numpy as np
```

```python
ref_em_prod = 10800

ref_em_storage = 132

fin_em_prod = 9494
fin_ore_prod = 3797
fin_crop_prod = 0
fin_wood_prod = 0
fin_fish_prod = 0

fin_em_storage = 116
fin_ore_storage = 66
fin_crop_storage = 22
fin_wood_storage = 22
fin_fish_storage = 18
```

```python
min_time_passed = (ref_em_storage / ref_em_prod * 60 * 60) % 60
max_time_passed = (ref_em_storage / ref_em_prod * 60 * 60 + 3) % 60
print(min_time_passed, max_time_passed)
```

```python
def usage_guess(time_passed, fin_storage, fin_prod):
    generation_snapshot = time_passed * fin_prod/60/60
    hourly_usage = (-generation_snapshot + fin_storage) / (60-time_passed) * 60 * 60
    return hourly_usage
```

```python
print("Min Em Prediction: ", usage_guess(min_time_passed, fin_em_storage, fin_em_prod))
print("Min Ore Prediction: ", usage_guess(min_time_passed, fin_ore_storage, fin_ore_prod))
print("Min Crop Prediction: ", usage_guess(min_time_passed, fin_crop_storage, fin_crop_prod))
print("Min Wood Prediction: ", usage_guess(min_time_passed, fin_wood_storage, fin_wood_prod))
print("Min Fish Prediction: ", usage_guess(min_time_passed, fin_fish_storage, fin_fish_prod))
print("---")
print("Max Em Prediction: ", usage_guess(max_time_passed, fin_em_storage, fin_em_prod))
print("Max Ore Prediction: ", usage_guess(max_time_passed, fin_ore_storage, fin_ore_prod))
print("Max Crop Prediction: ", usage_guess(max_time_passed, fin_crop_storage, fin_crop_prod))
print("Max Wood Prediction: ", usage_guess(max_time_passed, fin_wood_storage, fin_wood_prod))
print("Max Fish Prediction: ", usage_guess(max_time_passed, fin_fish_storage, fin_fish_prod))
```
